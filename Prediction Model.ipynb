```python
import numpy as np
import pandas as pd
from tensorflow import keras
import sklearn
```

```python
dataset = pd.read_csv('8 Cities Previous Rainfall Data.csv')
```

```python
dataset.head()
```

```python
dataset.corr()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
```

```python
one_h=ohe.fit_transform(dataset[["City"]]).toarray()
```

```python
one_hd=pd.DataFrame(one_h)
```

```python
dataset=dataset.drop("City",axis=1)
```

```python
X = dataset.drop("PRECIPITATION",axis=1)
```

```python
y = dataset["PRECIPITATION"]
```

```python
y=np.array(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(X)
```

```python
sd_data=sc.transform(X)
```

```python
sd_data = pd.DataFrame(sd_data)
```

```python
sd_data
```

```python
dropped=pd.concat([sd_data,one_hd],axis=1,ignore_index=True)
```

```python
dropped
```

```python
for i in range(66008):
    if y[i]<1:
        y[i]=0
    elif y[i]<15:
        y[i]=1
    elif y[i]<30:
        y[i]=2
    else:
        y[i]=3
```

```python
y=pd.DataFrame(y)
```

```python
y
```

```python
y.value_counts()
```

```python
dropped=dropped.drop(1,axis=1)
```

```python
X=dropped
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.8,shuffle=True)
```

```python
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5,shuffle=True)
```

```python
def build_model(n_h=1,n_n=30,lr=3e-3,input_shape=X_train.shape[1:]):
    ann = keras.models.Sequential()
    options = {"input_shape": input_shape}
    for layer in range(n_h):
        ann.add(keras.layers.Dense(n_n, activation="relu", **options))
        options = {}
    ann.add(keras.layers.Dense(4,activation="softmax", **options))
    optimizer = keras.optimizers.SGD(lr)
    ann.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer,metrics=['accuracy'])
    return ann
```

```python
keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
keras_reg.fit(X_train,y_train,epochs=100,
             validation_data=(X_valid,y_valid),
             callbacks = [keras.callbacks.EarlyStopping(patience=10)])
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_h": [0, 1, 2, 3],
    "n_n": np.arange(1, 100),
    "lr": reciprocal(3e-4, 3e-2),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3)
rnd_search_cv.fit(X_train, y_train, epochs=100,
validation_data=(X_valid,y_valid))
```

```python
rnd_search_cv.best_params_
```

```python
final_model = rnd_search_cv.best_estimator_.model
```

```python
final_model.summary()
```

```python
final_model.evaluate(X_test,y_test)
```

```python
final_model.save("model.h5")
```
